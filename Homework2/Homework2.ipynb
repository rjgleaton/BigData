```python
import pandas as pd
import csv, pdb
from sklearn import tree
```

```python
df = pd.read_csv("C:\\Users\\rjgle\\Documents\\CSCE 587\\Homework2\\fetal_health.csv")
df
```

```python
train_df = df.iloc[0:1500]
test_df = df.iloc[1500:]
```

```python
train_x = train_df[train_df.columns[0:-1]]
train_y = train_df['fetal_health']
```

```python
test_x = test_df[test_df.columns[0:-1]]
test_y = test_df['fetal_health']
```

```python
clf=tree.DecisionTreeClassifier()
clf.fit(train_x,train_y)
tree.plot_tree(clf)
```

```python
test_predict_Y=clf.predict(test_x)
test_predict_Y
```

```python
from sklearn.metrics import confusion_matrix
```

```python
import numpy as np
#Calculate confusion matrix items
cm = confusion_matrix(test_y, test_predict_Y)

FP = cm.sum(axis=0) - np.diag(cm)  
FN = cm.sum(axis=1) - np.diag(cm)
TP = np.diag(cm)
TN = cm.sum() - (FP + FN + TP)

print(cm)
print("False Positive Rates: ",FP)
print("False Negative Rates: ",FN)
print("True Positive Rates: ", TP)
print("True Negative Rates: ",TN)
```

```python
#Precision
from sklearn.metrics import precision_score

precision_score(test_y, test_predict_Y, average='micro')
```

```python
#Recall
from sklearn.metrics import recall_score

recall_score(test_y, test_predict_Y, average='micro')
```

```python
#F1 Score
from sklearn.metrics import f1_score

f1_score(test_y, test_predict_Y, average='micro')
```

```python
#MCC
from sklearn.metrics import matthews_corrcoef

matthews_corrcoef(test_y, test_predict_Y)
```

```python
adult_data = pd.read_csv('C:\\Users\\rjgle\\Documents\\CSCE 587\\Homework2\\adult.data', header=None)
adult_data
```

```python
train_x = adult_data[adult_data.columns[0:-1]]
train_y = adult_data[14]
features_to_encode = train_x.columns[train_x.dtypes==object].tolist()
features_to_encode
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.ensemble import RandomForestClassifier
```

```python
col_trans = make_column_transformer(
        (OneHotEncoder(handle_unknown="ignore"), features_to_encode), remainder = "passthrough")
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=12345)
```

```python
from sklearn.pipeline import make_pipeline
clf = make_pipeline(col_trans, rf_classifier)
```

```python
clf.fit(train_x, train_y)
```

```python
adult_test = pd.read_csv('C:\\Users\\rjgle\\Documents\\CSCE 587\\Homework2\\adult.test', header=None)

test_x = adult_test[adult_data.columns[0:-1]]
test_y = adult_test[14]

#Values have error in them by default
test_y = test_y.str.replace(" <=50K.", " <=50K", regex=False)
test_y = test_y.str.replace(' >50K.',' >50K', regex = False)
```

```python
#This looks similar to the curve in the towardsdatascience article that was provided so I will assume it's good
#ROC plot
from sklearn.metrics import plot_roc_curve
plot_roc_curve(clf, test_x, test_y)
plt.show()
```

```python
#AUC Score 
from sklearn.metrics import roc_auc_score

train_probs = clf.predict_proba(train_x)[:,1]
probs = clf.predict_proba(test_x)[:,1]

print(f'Train ROC AUC Score: {roc_auc_score(train_y, train_probs)}')
print(f'Test ROC AUC  Score: {roc_auc_score(test_y, probs)}')
```

```python
from sklearn.model_selection import GridSearchCV

param_grid={'randomforestclassifier__max_depth':[3,5,7,10],'randomforestclassifier__max_features':["auto","sqrt","log2"]}
clf_hyper = GridSearchCV(clf, param_grid, cv=10)

clf_hyper.fit(train_x, train_y)
```

```python
#Report best f1 score after hyper-paramter search
test_predict_Y=clf_hyper.predict(test_x)
f1_score(test_y, test_predict_Y, average='micro')
```
